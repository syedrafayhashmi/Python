```python
import os
from PIL import Image
import numpy as np
import PIL
```

```python
path = "R:\Module_2\Module_2_assignment\photos"
size = (200,200)
def resize_func(path,size):
    for fname in os.listdir(path):
        image = Image.open(os.path.join(path,fname))
        resized_img = image.resize(size, Image.ANTIALIAS)
        resized_img_name = os.path.join(path, "resized_" + fname)
        resized_img.save(resized_img_name,quality = 100)
        os.remove(path+"\\"+fname)
resize_func(path,size)
```
